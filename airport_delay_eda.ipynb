# US Flight Delays Exploratory Data Analysis

```python
data_filename = './sample_data/flights.csv'
```

## Data Preparation and Cleaning
- Load the file using pandas
- Look at information about file
- Fix any missing or incorrect values about the data or columns

```python
import pandas as pd
```

```python
df = pd.read_csv(data_filename)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Drop some columns that are not irrelevant to my analysis
columns_to_drop = ['WHEELS_ON', 'WHEELS_OFF', 'TAXI_IN', 'TAXI_OUT', 'CANCELLATION_REASON', 'TAIL_NUMBER']
df = df.drop(columns=columns_to_drop)
```

```python
# Go through all the different types of delays and change 'NaN' to 0
delays = ['WEATHER_DELAY', 'LATE_AIRCRAFT_DELAY', 'AIRLINE_DELAY', 'SECURITY_DELAY', 'AIR_SYSTEM_DELAY','ARRIVAL_DELAY','DEPARTURE_DELAY']
df[delays] = df[delays].fillna(0)
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# Now I will drop all rows that have missing values (Since it is only ~95k flights it wont affect my analysis to simply drop them as they have not been properly documented)

df = df.dropna(subset=['DEPARTURE_TIME', 'AIR_TIME', 'ELAPSED_TIME', 'ARRIVAL_TIME'])
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head(20)
```

## Exploratory Analysis and Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Basic Distributions

- We can see from these histograms that flights rarely ever depart earlier than schedule, however they quite often reach their destination early

```python
sns.histplot(df['ARRIVAL_DELAY'], bins=100, kde=True)
plt.xlim(-50, 150)
plt.title('Arrival Delay Distribution')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Number of Flights')
plt.show()
```

```python
sns.histplot(df['DEPARTURE_DELAY'], bins=100, kde=True)
plt.xlim(-50, 150)
plt.title('Departure Delay Distribution')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Number of Flights')
plt.show()
```

### Flights Over Time
- The monthly bar chart seems to suggest that seasons/months dont affect flight counts as the amount of flights each month are relatively equivalant
- Flights appear to be less frequent on saturday (and weekends in general)
- Most flights happen between 4AM - 10PM

```python
monthly_flights = df.groupby('MONTH').size()
```

```python
monthly_flights.plot(kind='bar', title="Number of flights per month")
plt.xlabel('Month')
plt.ylabel('Number of flights')
plt.show()
```

```python
weekly_flights = df.groupby('DAY_OF_WEEK').size()
```

```python
weekly_flights.plot(kind='bar', title="Number of flights on each day of the week")
plt.xlabel('Weekday')
plt.ylabel('Number of flights')
plt.show()
```

```python
df['DEP_HOUR'] = df['DEPARTURE_TIME'] // 100
hourly_flights = df.groupby('DEP_HOUR').size()
```

```python
hourly_flights.plot(kind='line', title='Number of Flights per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Flights')
plt.show()
```

# Airports
- There seems to be no correlation between busyness of an airport and its delay factors (Departure Delay and Arrival Delay)
- It appears to be the case that less busy airports struggle more with delay probably due to funding (Busier airports probably have better infrastructure)

```python
# Create subsets of all the origin and destination airports with their average delay

origin_airports = df.groupby('ORIGIN_AIRPORT').agg(
    num_flights=('ORIGIN_AIRPORT', 'size'),
    avg_dep_delay=('DEPARTURE_DELAY', 'mean')
)

destination_airports = df.groupby('DESTINATION_AIRPORT').agg(
    num_flights=('DESTINATION_AIRPORT', 'size'),
    avg_arr_delay=('ARRIVAL_DELAY', 'mean')
)
```

```python
origin_airports.shape, destination_airports.shape
```

```python
# Sort the airports by busiest

top_20_origin = origin_airports.sort_values(by='num_flights', ascending=False)
top_20_destination = destination_airports.sort_values(by='num_flights', ascending=False)
```

```python
plt.figure(figsize=(20, 8))
plt.scatter(top_20_origin['num_flights'], top_20_origin['avg_dep_delay'], color='blue')
plt.title('Origin Airports: Flights vs. Departure Delay')
plt.xlabel('Number of Flights')
plt.ylabel('Average Departure Delay (minutes)')
plt.xscale('log')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.scatter(top_20_destination['num_flights'], top_20_destination['avg_arr_delay'], color='blue')
plt.title('Destination Airports: Flights vs. Arrival Delay')
plt.xlabel('Number of Flights')
plt.ylabel('Average Arrival Delay (minutes)')
plt.xscale('log')
plt.grid(True)
plt.show()
```

# Delay Deep Dive

- There doesnt seem to be a correlation between flight distance and arrival delay
- Departure delay is at its greatest between 11PM - 4AM, which seems counter intuitive as an earlier graph showed that the least amount of fligths actually happen between those times (Under staffing issues perhaps?)

```python
# Plot Arrival Delay vs Flight Distance

plt.figure(figsize=(10, 6))
plt.scatter(df['DISTANCE'], df['ARRIVAL_DELAY'], alpha=0.5, color='purple')
plt.title('Arrival Delay vs Flight Distance')
plt.xlabel('Flight Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
```

```python
avg_dep_delay_per_hour = df.groupby('DEP_HOUR')['DEPARTURE_DELAY'].mean()

# Plotting the average departure delay per hour
plt.figure(figsize=(10, 6))
avg_dep_delay_per_hour.plot(kind='line', color='green', marker='o')
plt.title('Average Departure Delay per Scheduled Hour')
plt.xlabel('Scheduled Departure Hour')
plt.ylabel('Average Departure Delay (minutes)')
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```

### Interactive HeatMap

- You can see that some of the famous destination, such as los angeles, experience alot more air traffic.
- Something to note aswell is that the heatmaps of the destination vs origin are actually quite similir indicating that most flights are two way flights

```python
import folium
from folium.plugins import HeatMap
```

```python
# Load airports datafram
airports = pd.read_csv('./sample_data/airports.csv')
```

```python
airports.head()
```

```python
airports.isna().sum()
```

```python
airports = airports.dropna(subset=['LONGITUDE', 'LATITUDE'])
```

```python
airports.isna().sum()
```

```python
origin = df.groupby('ORIGIN_AIRPORT').size().reset_index(name='num_flights')
destination = df.groupby('DESTINATION_AIRPORT').size().reset_index(name='num_flights')
```

```python
is_numeric1 = origin['ORIGIN_AIRPORT'].str.isnumeric()
is_numeric2 = destination['DESTINATION_AIRPORT'].str.isnumeric()

origin = origin[~is_numeric1]
destination = destination[~is_numeric2]
```

```python
origin = origin.reset_index(drop=True)
destination = destination.reset_index(drop=True)
```

```python
origin = pd.merge(origin, airports[['IATA_CODE', 'LATITUDE', 'LONGITUDE']], left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', how='left')
destination = pd.merge(destination, airports[['IATA_CODE', 'LATITUDE', 'LONGITUDE']], left_on='DESTINATION_AIRPORT', right_on='IATA_CODE', how='left')
```

```python
origin.isna().sum()
```

```python
destination.isna().sum()
```

```python
origin = origin.dropna(subset=['LONGITUDE', 'LATITUDE', 'IATA_CODE'])
destination = destination.dropna(subset=['LONGITUDE', 'LATITUDE', 'IATA_CODE'])
```

```python
m_dep = folium.Map(location=[39.8283, -98.5795], zoom_start=5)
m_arr = folium.Map(location=[39.8283, -98.5795], zoom_start=5)
```

```python
heat_data = [
    [row['LATITUDE'], row['LONGITUDE'], row['num_flights']] for _, row in origin.iterrows()
]

heat_data = [
    [row['LATITUDE'], row['LONGITUDE'], row['num_flights']] for _, row in destination.iterrows()
]

HeatMap(heat_data).add_to(m_dep)
HeatMap(heat_data).add_to(m_arr)

m_dep.save('airport_heatmap.html')
m_arr.save('airport_heatmap.html')
```

## Conclusion

1. From the data, we see that flights rarely depart early, but early arrivals are more common. Departure delays peak between 11 PM and 4 AM, which is unusual given the lower number of flights during these hours—this could be due to staffing or operational issues.

2. Flight counts are fairly consistent throughout the year, and weekends, especially Saturdays, have fewer flights. Most flights happen between 4 AM and 10 PM.

3. Southwest Airlines (WN) is the most popular and reliable carrier, with lower delays compared to airlines like Spirit (NK) and Frontier (F9), which have fewer flights but higher delays.

4. There’s no clear link between flight distance and arrival delays, suggesting other factors (like weather or operations) are more important. Major airports, like Los Angeles, see more traffic, and heatmaps show that many flights are round trips between airports.

5. Overall, it seems that delays are more about airline operations and timing than just traffic or distance. There's room for improvement, especially in late-night flight handling and efficiency.
